```python
###Q1

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n = 30
m = 0.0
s = 1.0
x = norm(loc=m, scale=s).rvs(size=n)  # data

G = 10_000
theta = np.zeros(G)
tau   = np.zeros(G)

theta[0] = np.mean(x)
tau[0]   = 1.0

for g in range(G - 1):
    
    mean_theta = np.mean(x)
    var_theta  = 1.0 / (n * tau[g])
    theta[g+1] = np.random.normal(mean_theta, np.sqrt(var_theta))
    
    sse   = np.sum((x - theta[g+1])**2)
    alpha = n / 2.0
    beta  = sse / 2.0
    tau[g+1] = np.random.gamma(alpha, 1.0/beta) 

burn_in = 500
fig, axs = plt.subplots(1, 2, figsize=(12,4))

axs[0].plot(theta[burn_in:], 'b-', alpha=0.5)
axs[0].set_title(r'$\theta$ trace')
axs[0].set_xlabel('Iteration')
axs[0].set_ylabel(r'$\theta$')

axs[1].plot(tau[burn_in:], 'r-', alpha=0.5)
axs[1].set_title(r'$\tau$ trace')
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel(r'$\tau$')

plt.tight_layout()
plt.show()
```

```python
###Q2

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n = 30
true_theta = 0.0
true_tau   = 1.0

x = norm(loc=true_theta, scale=1/np.sqrt(true_tau)).rvs(size=n)


theta0  = 2.0
tau0_sq = 1.0
sigma_sq = 2.0
lam      = 2.0

alpha0 = sigma_sq / 2
beta0  = lam / 2

G = 10000
theta_samples = np.zeros(G)
tau_samples   = np.zeros(G)

theta_samples[0] = np.mean(x)
tau_samples[0]   = 1.0


for g in range(G - 1):
    tau_g = tau_samples[g]
    
    var_theta = 1.0 / ((1.0/tau0_sq) + n * tau_g)
    mean_theta = var_theta * ((theta0/tau0_sq) + tau_g * np.sum(x))
    
    theta_samples[g+1] = np.random.normal(mean_theta, np.sqrt(var_theta))
    
    sse = np.sum((x - theta_samples[g+1])**2)
    
    alpha_post = alpha0 + n/2.0
    beta_post  = beta0  + 0.5 * sse
    
    tau_samples[g+1] = np.random.gamma(shape=alpha_post, scale=1.0/beta_post)


burn_in = 1000

theta_chain = theta_samples[burn_in:]
tau_chain   = tau_samples[burn_in:]

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].plot(theta_chain, alpha=0.6)
axs[0].set_title(r'$\theta$ Trace')
axs[0].set_xlabel('Iteration')
axs[0].set_ylabel(r'$\theta$')

axs[1].plot(tau_chain, alpha=0.6, color='red')
axs[1].set_title(r'$\tau$ Trace')
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel(r'$\tau$')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.hist(theta_chain, bins=50, density=True, alpha=0.7)
plt.title("Posterior of theta")

plt.subplot(1,2,2)
plt.hist(tau_chain, bins=50, density=True, alpha=0.7, color='red')
plt.title("Posterior of tau")

plt.tight_layout()
plt.show()
```
